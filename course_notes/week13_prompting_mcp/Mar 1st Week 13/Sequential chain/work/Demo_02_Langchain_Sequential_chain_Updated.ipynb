#__Desciption__:
In this activity, let's explore how to use LangChain’s Sequential Chain functionality. This tutorial is designed to demonstrate the use of LangChain’s Sequential Chain in the context of a hypothetical TV show called **Banking Chronicles**.

# **Step 1: Import the Necessary Modules**

```python
from langchain_openai import ChatOpenAI
from langchain.prompts import PromptTemplate
from langchain.schema.runnable import RunnableSequence
import textwrap
```

# **Step 2: Define a Function to Print Responses from Our Chains**


*   This function is used throughout the tutorial to display the outputs of our chains.

```python
def display_response(response: str):
    print("\n".join(textwrap.wrap(response, width=100)))
```

# **Step 3: Initialize the Chat Model**


*   This model generates the responses for our chains.

```python

chat_model = ChatOpenAI(model="gpt-3.5-turbo", temperature=0)
```

# **Step 4: Define the First Chain**


*   Define the first chain, which generates a scenario for a new episode based on a given theme.

```python
scenario_template = """
You have to come up with a scenario (along with a 20–50 word description)
for a new episode of the TV show "Banking Chronicles" based on the theme

{theme_input}

ANSWER:
"""

scenario_prompt = PromptTemplate(
    input_variables=["theme_input"],
    template=scenario_template
)

# Combine prompt + model into a runnable chain
scenario_chain = scenario_prompt | chat_model
```

# **Step 5: Define the Second Chain**


*   Define the second chain, which generates a dialogue based on the scenario created by the first chain.

```python
dialogue_template = """
Generate a short dialogue between the bank manager and a customer
from the TV show "Banking Chronicles" for a new episode based on the scenario

{scenario}

ANSWER:
"""

dialogue_prompt = PromptTemplate(
    input_variables=["scenario"],
    template=dialogue_template
)

dialogue_chain = dialogue_prompt | chat_model
```

# **Step 6: Define the Sequential Chain**



*   Define the Sequential Chain, which chains the scenario and dialogue together.

```python
chain = RunnableSequence(scenario_chain | dialogue_chain)
```

# **Step 7: Run the Sequential Chain**


*   Run the Sequential Chain with the **Investment Strategies** theme and print the resulting conversation.

```python
theme = "Investment Strategies"  # you can take this as user input if needed
result = chain.invoke({"theme_input": theme})

print("\n Generated Episode Script:")
display_response(result.content)
```

```python
from langchain.chains import LLMChain, SimpleSequentialChain
scenario_chain = LLMChain(llm=chat_model, prompt=scenario_prompt)
dialogue_chain = LLMChain(llm=chat_model, prompt=dialogue_prompt)
sq = SimpleSequentialChain(chains=[scenario_chain ,dialogue_chain], verbose=True)
print(sq.run({"input":"comedy"}))
```

```python
scenario_template = """
You have to come up with a scenario (along with a 20–50 word description)
for a new episode of the TV show "Banking Chronicles" based on the theme

{theme_input}

ANSWER:
"""

scenario_prompt = PromptTemplate(
    input_variables=["theme_input"],
    template=scenario_template
)
first_chain = LLMChain(llm=chat_model, prompt=scenario_prompt, output_key="scenarioooo")
```

```python
dialogue_template = """
Generate a short dialogue between the bank manager and a customer
from the TV show "Banking Chronicles" for a new episode based on the scenario
for the {audience}
{scenarioooo}

ANSWER:
"""

dialogue_prompt = PromptTemplate(
    input_variables=["scenarioooo", "audience"],
    template=dialogue_template
)
second_chain = LLMChain(llm=chat_model, prompt=dialogue_prompt, output_key="dialogue")
```

```python
from langchain.chains import SequentialChain
overall = SequentialChain(chains=[first_chain, second_chain],
               input_variables=["theme_input", "audience"],
               output_variables=["scenarioooo", "dialogue"], verbose=True)
```

```python
overall.invoke({"theme_input":"Reality show",
               "audience":"students"})
```

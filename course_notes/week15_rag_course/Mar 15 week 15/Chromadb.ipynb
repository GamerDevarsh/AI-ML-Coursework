```python
import numpy as np
embed = np.array(embed).astype('float32')
```

```python
dimension = embed.shape[1]
```

```python
eu = faiss.IndexFlatL2(dimension)
eu.add(embed)
```

```python
faiss.write_index(eu, "eu_index")
```

```python
i1 = faiss.read_index("eu_index")
```

```python
query = "can we get top swimmers name"
```

```python
q_emb = model.encode([query]).astype("float32")
```

```python
distances, indices = i1.search(q_emb, k=2)
```

```python
quantizer = faiss.IndexFlatIP(dimension)
clustering = faiss.IndexIVFFlat(quantizer, dimension, 10)
clustering.train(embed)
clustering.add(embed)
```

```python
D, I = clustering.search(q_emb, k=2)
```

```python
df["combined"].iloc[I[0]]
```

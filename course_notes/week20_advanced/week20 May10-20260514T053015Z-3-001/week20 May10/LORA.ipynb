```python
device = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
model_name = "gpt2"
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name).to(device)
```

```python
def get_output(model, data, max_new_tokens=100):
    input_ids = tokenizer(data, return_tensors="pt").to(device)
    output = model.generate(**input_ids, max_new_tokens=max_new_tokens)
    return tokenizer.decode(output[0])
```

```python
query = "I want you to act as a motivational coach"
```

```python
get_output(model, query)
```

```python
tokenizer.pad_token = tokenizer.eos_token
data_token = data.map(lambda x: tokenizer(x["prompt"], truncation=True, padding=True,max_length=360), batched=True)
```

```python
print(data_token["train"][2])
```

```python
from peft import LoraConfig, get_peft_model
```

```python
lora_config = LoraConfig(r=64,
           lora_alpha=1,#2*r
           target_modules=['c_attn', 'c_proj'],
           lora_dropout=0.05,
           bias="lora_only",
           task_type="CAUSAL_LM")
```

```python
# !pip install --upgrade torchao
```

```python
peft_model = get_peft_model(model, lora_config)
```

```python
peft_model.print_trainable_parameters()
```

```python
from transformers import TrainingArguments, Trainer
```

```python
args = TrainingArguments(output_dir="peft",
                  auto_find_batch_size=False,
                  num_train_epochs=6,
                  per_device_train_batch_size=10,
                  learning_rate=0.005)
```

```python
data_token["train"]
```

```python
import transformers
```

```python
tr = Trainer(args=args,
        model=peft_model,
        train_dataset=data_token["train"],
             data_collator=transformers.DataCollatorForLanguageModeling(tokenizer, mlm=False))
```

```python
tokenizer.pad_token = tokenizer.eos_token
tr.train()
```

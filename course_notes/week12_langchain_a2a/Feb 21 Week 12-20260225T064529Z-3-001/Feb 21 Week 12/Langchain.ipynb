```python
class callback123(BaseCallbackHandler):
    def on_chain_Start(self, serialized, inputs, **kwargs):
        print(f"[Chain_start] inputs={inputs}")
    
    def on_llm_Start(self, serialized, prompt, **kwargs):
        print(f"[llm_start] prompt={prompt}")
        
    def on_llm_end(self, serialized, **kwargs):
        print(f"[llm_end] text={response}")
        
    def on_chain_end(self, output, **kwargs):
        print(f"[chain_end] output={output}")
```

```python
from langchain.chains import SequentialChain
```

```python
scenario = """you have to come up wih a scenario {theme_input}
Answer"""

p1 = PromptTemplate(template=scenario, input_variables=["theme_input"])
first_chain = LLMChain(llm=llm, prompt=p1, output_key="scenariotheme")
```

```python
dialogue = """generate dialogue for a tv show, for the {audience}
{scenariotheme}
Answer"""
```

```python
p2 = PromptTemplate(input_variables=["scenariotheme", "audience"], template=dialogue)
second_chain = LLMChain(llm=llm, prompt=p2, output_key="dialoguesummary")
```

```python
overall_chain = SequentialChain(chains=[first_chain, second_chain],
                input_variables=["theme_input", "audience"],
                output_variables=["scenariotheme", "dialoguesummary"], verbose=True)
```

```python
%%time
response = overall_chain.invoke({"theme_input":"reality show",
                      "audience":"students"})
```

```python
print(response["dialoguesummary"])
```

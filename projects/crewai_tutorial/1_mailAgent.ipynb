**TOOL CALLING EXAMPLE**

```python
original_email = """
looping in Priya. TAS and PRX updates are in the deck. ETA for SDS integration is Friday.
Let's sync up tomorrow if SYNCBOT allows рҹҳҠ. ping me if any blockers.
"""
```

```python
from crewai.tools import BaseTool

class ReplaceJargonTool(BaseTool):
    name: str = "Replace Jargons"
    description: str = "Replaces jargon and abbreviations with more speciific terms"

    def _run(self, email: str) -> str:
     replacements = {
        "PRX": "Project Phoenix (internal AI revamp project)",
        "TAS": "technical architecture stack",
        "DBX": "client database cluster",
        "SDS": "Smart Data Syncer",
        "SYNCBOT": "internal standup assistant bot",
        "WIP": "in progress",
        "POC": "proof of concept",
        "ping": "reach out"
     }

     suggestions = []
     email_lower = email.lower()

     for jargon, replacement in replacements.items():
        if jargon.lower() in email_lower:
            suggestions.append(f"Consider replacing '{jargon}' with '{replacement}'")
     return "\n".join(suggestions) if suggestions else "No jargon found. Email looks good!"
    
jt = ReplaceJargonTool()
jt.run(original_email) #use this for testing the tool independently
```

```python
from crewai import Agent, Task, Crew

email_assistant = Agent(
    role="Email Assistant Agent",
    goal="Improve emails and make them sound professional and clear",
    backstory="A highly experienced communication expert skilled in professional email writing",
    verbose =True,
    llm=llm,
    tools=[jt]
)

email_task = Task(
    description=f"""Take the following rough email and rewrite it into a professional and polished version.
    Expand abbreviations and replace jargon with more specific terms:
    '''{original_email}'''""",
    agent=email_assistant,
    expected_output="A professional written email with proper formatting and content.",
)

crew = Crew(
    agents=[email_assistant],
    tasks=[email_task],
    verbose=True
)

result = crew.kickoff()
print(result)
```

**CREW OF AGENTS**

```python
from crewai import Agent, Task, Crew
```

```python
research_agent = Agent(
    role="Research Specialist",
    goal="Research interesting facts about the topic: {topic}",
    backstory="You are an expert at finding relevant and factual data.",
    verbose=True,
    llm=llm
)

writer_agent = Agent(
    role="Creative Writer",
    goal="Write a short blog summary using the research",
    backstory="You are skilled at writing engaging summaries based on provided content.",
    llm=llm,
    verbose=True,
)
```

```python
task1 = Task(
    description="Find 3-5 interesting and recent facts about {topic}.",
    expected_output="A bullet list of 3-5 facts",
    agent=research_agent,
)

task2 = Task(
    description="Write a 100-word blog post summary about {topic} using the facts from the research.",
    expected_output="A blog post summary",
    agent=writer_agent,
    context=[task1],
)
```

```python
crew = Crew(
    agents=[research_agent, writer_agent],
    tasks=[task1, task2],
    verbose=True
)
```

```python
crew.kickoff(inputs={"topic": "Artificial Intelligence"})
```

**CREW WITH TOOLS from crew ai** https://docs.crewai.com/en/concepts/tools

```python
from dotenv import load_dotenv

load_dotenv()
```

```python
from crewai import LLM
```

```python
llm =LLM(
    model="gemini/gemma-4-31b-it",
    temperature=0.1
)
```

```python
from crewai import Agent, Task, Crew
from crewai_tools import SerperDevTool 
```

```python
research_agent = Agent(
    role="Research Specialist",
    goal="Research interesting facts about the topic: {topic}",
    backstory="You are an expert at finding relevant and factual data.",
    tools=[SerperDevTool()],
    verbose=True,
    llm=llm
)

writer_agent = Agent(
    role="Creative Writer",
    goal="Write a short blog summary using the research",
    backstory="You are skilled at writing engaging summaries based on provided content.",
    llm=llm,
    verbose=True,
)
```

```python
task1 = Task(
    description="Find 3-5 interesting and recent facts about {topic} as of year 2025.",
    expected_output="A bullet list of 3-5 facts",
    agent=research_agent,
)

task2 = Task(
    description="Write a 100-word blog post summary about {topic} using the facts from the research.",
    expected_output="A blog post summary",
    agent=writer_agent,
    context=[task1],
)
```

```python
crew = Crew(
    agents=[research_agent, writer_agent],
    tasks=[task1, task2],
    verbose=True
)
```

```python
output = crew.kickoff(inputs={"topic": "The Future of electric vehicles"})
print(output)
```
